# This notebook draw some visualizations of the dataset focusing on the distribution of flight delay as it is the target feature

## Importing the need packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
# Set option to display all columns
pd.set_option('display.max_columns', None)

# Set the aesthetic style of the plots
sns.set(style='whitegrid')
```

## Importing the dataset

```python
# Get the current script's directory
current_script_dir = os.getcwd()

# Move up to the parent directory
parent_dir = os.path.dirname(current_script_dir)

# Define the sibling data directory name
data_dir = os.path.join(parent_dir, 'data')

# Use os.path.join to create the full file path
dataset_filename = "prosessed_dataset_2022.csv"
file_path = os.path.join(data_dir, dataset_filename)

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
display(df)
```

```python
# Create a 1x2 subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Make histogram that show the distribution of DEP_DELAY
sns.histplot(df['DEP_DELAY'], bins=len(df['DEP_DELAY'].unique()), ax=axes[0])
axes[0].set_title('Histogram of Departure Delays')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Frequency')

# Make histogram that show the distribution of DEP_DELAY_GROUP
sns.histplot(df['DEP_DELAY_GROUP'], bins=len(df['DEP_DELAY_GROUP'].unique()), ax=axes[1])
axes[1].set_title('Histogram of Departure Delay Groups')
axes[1].set_xlabel('Values')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a 1x2 subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Make histogram that show the distribution of DEP_DELAY
sns.histplot(df[df['DEP_DELAY'] <= 60]['DEP_DELAY'], bins=len(df[df['DEP_DELAY'] <= 60]['DEP_DELAY'].unique()), ax=axes[0])
axes[0].set_title('Histogram of Departure Delays')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Frequency')

# Make histogram that show the distribution of DEP_DELAY_GROUP
sns.histplot(df[df['DEP_DELAY_GROUP'] <= 4]['DEP_DELAY_GROUP'], bins=len(df[df['DEP_DELAY_GROUP'] <= 4]['DEP_DELAY_GROUP'].unique()), ax=axes[1])
axes[1].set_title('Histogram of Departure Delay Groups')
axes[1].set_xlabel('Values')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
